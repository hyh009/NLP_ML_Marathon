### 作業目的: 了解何謂Bias-Variance Tradeoff

### Q1: 一般模型誤差可拆為三部分，這三部分的誤差分別是？

Answer: 
1.   Bias error
2.   Variance error
1.   The noise 




### Q2: 請解釋何謂Bias-Variance Tradeoff

Answer:
### Bias error和Variance error間的平衡

在訓練模型時必須注意bias error 和 variance error間的平衡<br>
如果High bias & Low variance代表模型underfitting<br>
如果Low bias & High variance代表模型overfitting<br>
而最好的狀態則是Low bias & Low variance<br>
=>在訓練模型時可以透過觀察training data 和 testing data 套入模型時的分數來判斷是否有overfitting等問題
